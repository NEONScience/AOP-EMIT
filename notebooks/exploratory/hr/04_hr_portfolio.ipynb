## Introduction:
This project seeks to answer two questions:
1. How can Python be used to compare mineral distribution maps from the Earth Surface Mineral Dust Source Investigation (EMIT) instrument with classified surface reflectance data gathered by the NEON Airborne Observation Platform (AOP)?
2. What are the differences in mineral distribution and surface reflectance between burned and unburned areas? How does fire alter surface minerals?

The area of interest is the [Soaproot Saddle (SOAP) NEON field site](https://www.neonscience.org/field-sites/soap) in the Sierra National Forest in California. The SOAP site covers 1440 acres and has an average elevation of 1,210 m. The full elevation range is 1,044 m-1373 m. The land cover of SOAP is primarily evergreen forest and shrub/scrub. More specifically, the forest is mixed conifer and has steep hills with narrow drainages. SOAP's geology is granodiorite and quartz monzonite and there are a few different primary soil series: Holland, Chaix, Chawanakee, and Tollhouse.



*The national forest boundary in the plot above came from [this FS National Forests Dataset (US Forest Service Proclaimed Forests) dataset](https://data-usfs.hub.arcgis.com/datasets/usfs::fs-national-forests-dataset-us-forest-service-proclaimed-forests/about)*

Additionally, to answer the second question, two fires covering at least part of the SOAP site will be analyzed:
1. [Creek Fire](https://www.fire.ca.gov/incidents/2020/9/4/creek-fire/): started on September 4, 2020 and burned 379,895 acres
2. [Blue Fire](https://www.fire.ca.gov/incidents/2021/6/30/blue-fire/): started on June 30, 2021 and burned 277 acres.

These fires were chosen primarily because there is NEON surface reflectance data from before and after these fires. Also, at the SOAP site, these fires occurred the closest to when EMIT started gathering data.

## Methods:
An overview:
1. Set up our coding environment to work with NEON and EMIT data. (not on workflow diagram)
2. Identify co-located EMIT L2B Estimated Mineral Identification data AND NEON Airborne Observation Platform Flight Boundary data. Co-location: SOAP site and fire perimeters. (not on our workflow diagram)
3. Download and wrangle/clip the co-located EMIT and NEON data. These data will be used to create EMIT Mineral Distriubtion maps specifically where the SOAP AOP flight box boundaries are *and* where the fire perimeters were. The NEON surface reflectance data will be classified via an unsupervised clustering algorithm. The NEON Site Management data will be used more qualitatively to create site and fire descriptions.
4. Compare the maps and classified data to explore how fire disturbances affect Earth surface minerals.
5. End products: a reproducible Python tutorial, the aforementioned maps and classified data, and a comparison analysis between EMIT and NEON data.
